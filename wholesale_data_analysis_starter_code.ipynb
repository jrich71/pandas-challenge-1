## Part 2: Transform the Data
Do we know that this client spent the more money than client 66037? If not, how would we find out? Transform the data using the steps below to prepare it for analysis.

```python
# Create a column that calculates the subtotal for each line using the unit_price and the qty
clients_df["line_subtotal"] = clients_df["unit_price"] * clients_df["qty"]
clients_df[["unit_price", "qty", "line_subtotal"]]
```

```python
# Create a column for shipping price.
# Assume a shipping price of $7 per pound for orders over 50 pounds and $10 per pound for items 50 pounds or under.
def calculate_shipping(row):
    unit_weight = row["unit_weight"]
    qty = row["qty"]
    if unit_weight * qty > 50:
        shipping = (unit_weight * qty) * 7
    else:
        shipping = (unit_weight * qty) * 10
    return shipping

# Apply the function
clients_df["shipping_price"] = clients_df.apply(calculate_shipping, axis=1)

clients_df

```

```python
# Create a column for the total price using the subtotal and the shipping price along with a sales tax of 9.25%
clients_df["line_price"] = round((clients_df["line_subtotal"] + clients_df["shipping_price"]) * 1.0925, 2)
clients_df
```

```python
# Create a column for the cost of each line using unit cost, qty, and
# shipping price (assume the shipping cost is exactly what is charged to the client).
clients_df["line_cost"] = (clients_df["unit_cost"] * clients_df["qty"]) + clients_df["shipping_price"]

clients_df

```

```python
# Create a column for the profit of each line using line cost and line price
clients_df["line_profit"] = round(clients_df["line_price"] - clients_df["line_cost"], 2)

clients_df
```

## Part 4: Summarize and Analyze
Use the new columns with confirmed values to find the following information.

```python
# How much did each of the top 5 clients by quantity spend (i.e., line_price)? Check your work from Part 1 for client ids.

# Using GroupBy in order to separate the data into fields according to "client_id" values
top_buyers = clients_df.groupby(["client_id"])
top_five_spenders = top_buyers["line_price"].sum().sort_values(ascending=False).loc[top_5_clients]
top_five_spenders
```

```python
# Create a summary DataFrame showing the totals for the top 5 clients (spenders) with the following information:
# total units purchased, total shipping price, total revenue, and total profit.

# Filter the list
data = [33615, 66037, 46820, 38378, 24741]
filtered_clients = clients_df[clients_df["client_id"].isin(data)]

# Group by client ID and calculate totals
summary_df = filtered_clients.groupby("client_id").agg(
    total_units_purchased = ("qty", "sum"),
    total_shipping_price = ("shipping_price", "sum"),
    total_revenue = ("line_price", "sum"),
    total_profit = ("line_profit", "sum")
    ).reset_index()

# Display the summary DataFrame
summary_df
```

```python
# Format the data and rename the columns to names suitable for presentation.

presentation_df = summary_df.rename(columns={"client_id": "Client ID", 
                                             "total_units_purchased": "Total Units Purchased",
                                             "total_shipping_price": "Total Shipping Price",
                                             "total_revenue": "Total Revenue",
                                             "total_profit": "Total Profit"                                
                                             }
                                             )

# Define the money columns. 
presentation_df.info() # I see that the "money" columns (n=3) are floats. Is this what it means to define a "money column"?
```

```python
# Define a function named currency_format_millions that converts a dollar amount to millions.
def currency_format_millions(x):
    amount_millions = x / 1000000
    return amount_millions

# Apply the currency_format_millions function to only the money columns. 
presentation_df[["Amount1 in Millions", 
                 "Amount2 in Millions", 
                 "Amount3 in Millions"]] = presentation_df[["Total Shipping Price", 
                                                            "Total Revenue", 
                                                            "Total Profit"]].apply(currency_format_millions)

# Rename the columns to reflect the change in the money format.
presentation_df = presentation_df.rename(columns={"Amount1 in Millions": "Total Shipping Price ($ Millions)", 
                                                  "Amount2 in Millions": "Total Revenue ($ Millions)", 
                                                  "Amount3 in Millions": "Total Profit ($ Millions)"})

# drop duplicates
presentation_df = presentation_df.loc[:, ~presentation_df.columns.duplicated()]

# Display the dataframe
presentation_df
```

```python
# Sort the updated data by "Total Profit (millions)" form highest to lowest and assign the sort to a new DatFrame.

final_df = presentation_df.sort_values("Total Profit ($ Millions)", ascending=False)

final_df
```
